prime not prime

```python
num=10
count=0
for i in range(1,num+1):
  if num%i==0:
    count+=1
if count==2:
  print("prime")
else:
  print("not prime")
```

```python
num=11
for i in range(2,num):
  if num%i==0:
    print("not prime")
    break
else:
  print("prime")
```

```python
arr=[1,2,3,2,9,4,3]
result = list(set(arr))
print(result)
#find dupl
duplicates = []
for i in arr:
  if arr.count(i) > 1 and i not in duplicates:
    duplicates.append(i)
print(duplicates)
```

```python
arr = [5,2,8,1]

for i in range(len(arr)):
   for j in range(i+1, len(arr)):
       if arr[i] > arr[j]:
           arr[i], arr[j] = arr[j], arr[i]

print(arr)
```
